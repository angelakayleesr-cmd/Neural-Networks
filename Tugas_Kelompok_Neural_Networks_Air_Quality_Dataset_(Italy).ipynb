```python
# =====================================================
# 1. DOWNLOAD & EXTRACT DATASET
# =====================================================
!wget https://archive.ics.uci.edu/static/public/501/beijing+multi+site+air+quality+data.zip
!unzip -o beijing+multi+site+air+quality+data.zip
!unzip -o PRSA2017_Data_20130301-20170228.zip

# =====================================================
# 2. IMPORT LIBRARY
# =====================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# =====================================================
# 3. LOAD DATASET
# (contoh 1 stasiun: Aotizhongxin)
# =====================================================
df = pd.read_csv("PRSA_Data_20130301-20170228/PRSA_Data_Aotizhongxin_20130301-20170228.csv")

# =====================================================
# 4. DATA UNDERSTANDING & PREPARATION
# =====================================================

# Gabungkan kolom waktu
df['datetime'] = pd.to_datetime(df[['year','month','day','hour']])
df = df.sort_values('datetime')
df = df.set_index('datetime')

# Pilih fitur penting
features = [
    'PM2.5','PM10','SO2','NO2','CO','O3',
    'TEMP','PRES','DEWP','WSPM'
]

df = df[features]

# Tangani missing value
df = df.interpolate()

# =====================================================
# 5. NORMALISASI DATA
# =====================================================
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)

# =====================================================
# 6. MEMBUAT DATA TIME-SERIES
# =====================================================
def create_sequences(data, window_size=24):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size, 0])  # target PM2.5
    return np.array(X), np.array(y)

X, y = create_sequences(scaled_data, 24)

print("Shape X:", X.shape)
print("Shape y:", y.shape)

# =====================================================
# 7. SPLIT DATA (70:30)
# =====================================================
split = int(0.7 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# =====================================================
# 8. DESAIN NEURAL NETWORK (LSTM)
# =====================================================
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(X.shape[1], X.shape[2])),
    Dropout(0.2),
    LSTM(32),
    Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

model.summary()

# =====================================================
# 9. TRAINING MODEL
# =====================================================
history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.1,
    verbose=1
)

# =====================================================
# 10. EVALUASI MODEL
# =====================================================
loss, mae = model.evaluate(X_test, y_test)
print("Test MAE:", mae)

# =====================================================
# 11. VISUALISASI HASIL PREDIKSI
# =====================================================
y_pred = model.predict(X_test)

plt.figure(figsize=(12,5))
plt.plot(y_test[:200], label='Aktual')
plt.plot(y_pred[:200], label='Prediksi')
plt.legend()
plt.title("Prediksi PM2.5 Menggunakan LSTM")
plt.xlabel("Waktu")
plt.ylabel("PM2.5 (scaled)")
plt.show()
```
